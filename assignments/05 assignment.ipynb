## Task

The [Koch snowflake](https://en.wikipedia.org/wiki/Koch_snowflake) is a fractal curve formed by starting with an equilateral triangle and recursively modifying each straight line segment. At each iteration, every segment is replaced by four segments, each 1/3 the length of the original, forming an equilateral "bump":

![The first seven iterations in animation](https://upload.wikimedia.org/wikipedia/commons/f/fd/Von_Koch_curve.gif)

The **perimeter length** of the Koch snowflake after $n$ iterations follows a recursive pattern:   

- The initial equilateral triangle (base case $n = 1$) has side length $s$, so the perimeter length is $3s$.
- In each iteration ($n > 1$), each line segment of length $s$ is replaced by 4 segments of length $s/3$.

Your task is to write a **recursive function** that calculates the perimeter length of the Koch snowflake with initial side length $s$ after $n$ iterations.

**Comments and hints:**

- The function is defined for integer $n \ge 1$ and real $s \ge 0$. If some of these preconditions does not hold, the function should return `None`.
- If the input parameters are correct, the function should return a `float` or `int` number.
- **Do not use the `print` and `input` functions in this assignment.**
- Do not use any imports in this assignment.

```python
def koch_length(n, s):
     if type(n) != int or (type(s) != int and type(s) != float) or n < 1 or s < 0:
          return None
     if n == 1:
          return 3 * s
     else:
          return 4/3 * koch_length(n-1, s)
```
