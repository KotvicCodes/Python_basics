````python
#
import os

#! Process Temperatures Function
def process_temperatures(in_filename, out_filename, sort="avg"):
     fileExisted = os.path.exists(out_filename)

     #* Rewrite document into array
     try:
          with open(f"{in_filename}", "rt") as input:
               lines = input.readlines()
     except:
          return -1

     #* Clean & check for inconsistencies
     header = lines.pop(0)
     fixedLines = []

     for line in lines:
          data = line.split(",")

          if len(data) != 6:
               return -2
          
          try:
               data[0] = int(data[0])
               data[1] = int(data[1])
               data[2] = int(data[2])
               data[3] = float(data[3])
               data[4] = float(data[4])
               data[5] = float(data[5].replace("\n", ""))
          except:
               return -2

          if data[0] < 0 or data[1] < 0 or data[1] > 12 or data[2] < 0 or data[2] > 31:
               return -2

          fixedLines.append(data)

     #* Sort accordingly
     sortingIndex = -1

     if sort == "avg":
          sortingIndex = 3
     elif sort == "max":
          sortingIndex = 4
     elif sort == "min":
          sortingIndex = 5
     else:
          return -2
     
     fixedLines.sort(key=lambda x: x[sortingIndex])

     #* Build a string
     outputArray = []
     for line in fixedLines:
          lineString = str(line[0]) + "," + str(line[1]) + "," + str(line[2]) + "," + str(line[3]) + "," + str(line[4]) + "," + str(line[5])
          outputArray.append(lineString)
     outputString = header + "\n".join(outputArray)

     #* Write to output file
     try:
          with open(out_filename, "wt") as output:
               output.write(outputString)
               return 0
     except:
          if not fileExisted and os.path.exists(out_filename):
               os.remove(out_filename)
          return -1
````
